# Numpy Project :- Student Result Analysis

##### **Description About Data**
There is a 'students_marks.csv' file which contains data of 100 students.
There are 100 rows where every row representing a student data. 
There are four columns where first column contains roll number and 
next three columns contains marks of first, second and third subject respectively.

**Import the numpy library**

```python
import numpy as np
```

**Read the data from csv file into numpy array**

```python
students_data = np.loadtxt('students_marks.csv', delimiter=',')
```

**Find the shape of data**

```python
students_data.shape
```

**Display first 5 rows of data**

```python
print(students_data[:5])
```

**Find the total of marks of each student**

```python
total_marks = students_data[:,1:].sum(axis=1)
print(total_marks[:5])
```

**Add a new column that will store the total of marks of each student**

```python
print(total_marks.reshape(1, -1)[:,:5])
```

```python
print(total_marks.reshape(-1, 1)[:5])
```

```python
print(total_marks[:5])
```

```python
total_marks = total_marks.reshape(-1, 1)
```

```python
print(total_marks[:5])
```

```python
students_data = np.concatenate((students_data, total_marks), axis=1)
```

**Display first 5 rows of data with newly added column for total of marks**

```python
print(students_data[:5])
```

**Find the average of marks of each student**

```python
average_marks = students_data[:,-1] / 3
print(average_marks[:5])
```

**Add a new column that will store the average of marks of each student**

```python
average_marks = average_marks.reshape(-1, 1)
print(average_marks[:5])
```

```python
students_data = np.concatenate((students_data, average_marks), axis=1)
```

**Display first 5 rows of data with newly added column for average of marks**

```python
print(students_data[:5])
```

**Find the average of marks of third subject**

```python
print(f"Average Marks of third subject: {students_data[:,3].mean()}")
```

**Count how many number of students have scored more than or equal to 90 marks in first subject**

```python
print(f"Number of Students scored more than or equal to 90 in first subject: {
(students_data[:,1] >= 90).sum()}")
```

**Count how many number of students have scored more than or equal to 90 marks in second subject**

```python
print(f"Number of Students scored more than or equal to 90 in second subject: {
(students_data[:,2] >= 90).sum()}")
```

**Count how many number of students have scored more than or equal to 90 marks in third subject**

```python
print(f"Number of Students scored more than or equal to 90 in third subject: {
(students_data[:,3] >= 90).sum()}")
```

**Count how many number of students have scored average marks more than or equal to 75**

```python
print(f"Number of Students scored average marks more than or equal to 75: {
(students_data[:,-1] >= 75).sum()}")
```

**Count the number of students who have scored average marks less than 50**

```python
print(f"Number of Students scored average marks less than 50: {
(students_data[:,-1] < 50).sum()}")
```

**Count the number of students who have scored average marks between 80 and 90**

```python
print(f"Number of Students scored average marks between 80 and 90: {
((students_data[:,-1] >= 80) & (students_data[:,-1] >= 90)).sum()}")
```

**Find the students data who have scored marks more than or equal to 90 in all subjects**

```python
all_subjects_90 = students_data[(students_data[:,1] >= 90) & 
(students_data[:,2] >= 90) & (students_data[:,3] >= 90)]
print(all_subjects_90)
```

**Count how many number of students have scored marks more than or equal to 90 in all subjects**

```python
print(f"Number of Students scored marks more than or equal to 90 in all subject: {
all_subjects_90.shape[0]}")
```

**Find the students data who have scored marks more than or equal to 90 in any subject**

```python
any_subject_90 = students_data[(students_data[:,1] >= 90) | 
(students_data[:,2] >= 90) | (students_data[:,3] >= 90)]
print(any_subject_90)
```

**Count how many number of students have scored marks more than or equal to 90 in any subject**

```python
print(f"Number of Students scored marks more than or equal to 90 in any subject: {
any_subject_90.shape[0]}")
```

**Save the students data who have scored marks more than or equal to 90 in all subjects to 'all_subjets_90' csv file**

```python
np.savetxt("all_subjects_90.csv", all_subjects_90, delimiter=",", 
header="Roll Number,Marks 1,Marks 2,Marks 3,Total Marks,Average Marks",
fmt=["%d","%d","%d","%d","%d","%.2f"], comments="")
print("Data is saved successfully")
```

**Save the students data who have scored marks more than or equal to 90 in any subject to 'any_subjet_90' csv file**

```python
np.savetxt("any_subject_90.csv", any_subject_90, delimiter=",", 
header="Roll Number,Marks 1,Marks 2,Marks 3,Total Marks,Average Marks",
fmt=["%d","%d","%d","%d","%d","%.2f"], comments="")
print("Data is saved successfully")
```
